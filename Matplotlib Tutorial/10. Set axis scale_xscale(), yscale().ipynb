<h3> <strong> 1. x축 스케일 지정 </strong> </h3>

```python
import matplotlib.pyplot as plt
import numpy as np
 
x = np.linspace(-10, 10, 100) # -10~10까지 동일 간격으로 100개 원소 생성
y = x**3

plt.plot(x, y)
plt.show()
```

**xscale()** 함수를 사용해 x축의 스케일을 'symlog'로 지정해보자. <br>
symlog는 Symmetrical Log Scale의 약자로 축의 원점을 기준으로 양, 음의 방향이 대칭적인 로그 스케일로 표시된다.

```python
import matplotlib.pyplot as plt
import numpy as np
 
x = np.linspace(-10, 10, 100) # -10~10까지 동일 간격으로 100개 원소 생성
y = x**3

plt.plot(x, y)
plt.xscale('symlog')
plt.show()
```

<h3> <strong> 2. y축 스케일 지정 </strong> </h3>

yscale() 함수를 사용하면 y축의 스케일을 정할 수 있다. <br>
아래 코드는 각각 linear, log 스케일로 나타낸 그래프이다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 100)
y = np.exp(x)

plt.plot(x, y)
plt.yscale('linear')  # linear은 기본 default값의 결과와 동일하다.

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 100)
y = np.exp(x)

plt.plot(x, y)
plt.yscale('log')

plt.show()
```
