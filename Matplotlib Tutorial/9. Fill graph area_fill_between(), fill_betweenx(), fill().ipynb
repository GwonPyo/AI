<h3> <strong> 1. 기본 사용 </strong> </h3>

**fill_between() 함수**에 x[1:3], y[1:3]을 입력하면, (x[1], y[1]), (x[2], y[2]), (x[1], 0), (x[2], 0)을 잇는 영역이 채워진다.

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]; y = [2, 3, 5, 10]
plt.plot(x, y)
plt.fill_between(x[1:3], y[1:3], alpha=0.3) # alpha: 투명도 설정

plt.show()
```

**fill_betweenx() 함수**에 y[2:4], x[2:4]를 순서대로 입력하면, (x[2], y[2]), (x[3], y[3]), (0, y[2]), (0, y[3])을 잇는 영역이 채워진다.

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]; y = [2, 3, 5, 10]
plt.plot(x, y)
plt.fill_betweenx(y[2:4], x[2:4], alpha=0.7) 

plt.show()
```

<h3> <strong> 2. 두 그래프 사이 영역 채우기 </strong> </h3>

두 개의 그래프 사이 영역을 채우려면 두 개의 y값 리스트 y1, y2를 입력해주면 된다. <br>
아래 코드의 경우 (x[1], y[1]), (x[1], y[2]), (x[2], y[1]), (x[2], y[2]) 사이 영역을 채운다.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [2, 3, 5, 10]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.fill_between(x[1:3], y1[1:3], y2[1:3], color='lightgray', alpha=0.5)

plt.show()
```

<h3> <strong> 3. 다각형 영역 채우기 </strong> </h3>

**fill() 함수**에 x, y값의 리스트를 입력하면 x, y 점들로 정의되는 다각형 영역을 채울 수 있다

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [2, 3, 5, 10]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.fill([1.9, 1.9, 3.1, 3.1], [1.0, 4.0, 6.0, 3.0], color='lightgray', alpha=0.5)

plt.show()
```
