<h3> <strong> 1. 기본 사용 </strong> </h3>

그래프에 범례를 나타내기 위해서는 **plot() 함수**에 label 문자열을 지정해야 한다. <br>
그리고 이후 **legend() 함수**를 호출하면 된다.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()

plt.show()
```

<h3> <strong> 2. 위치 지정 </strong> </h3>

**loc 파이미터**를 숫자 쌍 튜플로 지정하면 해당 위치에 범례가 표시된다. <br>
(0.0, 0.0)는 데이터의 왼쪽 아래, (1.0, 1.0)은 데이터의 오른쪽 위를 의미한다. <br>
이외에도 'lower right'처럼 문자열로 지정할 수도 있다.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc=(0.5, 0.5))

plt.show()
```

<h3> <strong> 3. 열 개수 지정 </strong> </h3>

기본적으로 legend() 함수를 사용하여 여러 범례를 표시하면 아래처럼 하나의 행당 하나를 출력한다.

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc='best')

plt.show()
```

**ncol 파라미터**를 이용하면 하나의 행당 표시할 수 있는 범례의 개수를 늘려준다.

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc='best', ncol=2)

plt.show()
```

<h3> <strong> 4. 폰트 크기 지정 </strong> </h3>

**fontsize 파라미터**를 사용하면 폰트의 크기를 지정할 수 있다. 

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc='best', ncol=2, fontsize=14)

plt.show()
```

<h3> <strong> 5. 범례 테두리 꾸미기 </strong> </h3>

**frameon 파라미터**는 범례 텍스트 상자의 테두리를 표시할지 여부를 지정한다. <br>
**shadow 파라미터**는 텍스트 상자에 그림자를 표시해준다.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc='best', ncol=1, fontsize=10, frameon=False)

plt.show()
```
