<h3> <strong> 1. 기본 사용 </strong> </h3>

xlabel(), ylabel() 함수에 문자열을 입력하면 축에 레이블을 표시할 수 있다.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 4, 5])
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.show()
```

<h3> <strong> 2. 여백 지정 </strong> </h3>

xlabel(), ylabel() 함수의 **labelpad 파라미터**를 사용해 여백(padding)을 지정할 수 있다.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 4, 5])
plt.xlabel('X-Label', labelpad=15)
plt.ylabel('Y-Label', labelpad=20)
plt.show()
```

<h3> <strong> 3. 여백 지정 </strong> </h3>

xlabel(), ylabel() 함수의 **fontdict 파라미터**를 사용하면 축 레이블의 폰트 스타일을 지정할 수 있다.

```python
import matplotlib.pyplot as plt

font1 = {'family': 'serif',
         'color': 'b',
         'weight': 'bold',
         'size': 14
         }

font2 = {'family': 'fantasy',
         'color': 'deeppink',
         'weight': 'normal',
         'size': 'xx-large'
         }

# 위처럼 묶지 않으면 fontdict = {'family': 'serif','color': 'b','weight': 'bold','size': 14} 식으로 사용해야 한다.
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis', labelpad=15, fontdict=font1)
plt.ylabel('Y-Axis', labelpad=20, fontdict=font2)
plt.show()
```

<h3> <strong> 4. 위치 지정 </strong> </h3>

xlabel() 함수의 **loc 파라미터**로 레이블의 위치를 지정할 수 있다. <br>
ylabel() 함수도 동일하다.

단, 3.3 버전 이후부터 적용되었다.<br>
colab 코렙 기본 matplotlib은 3.2.2 버전이므로 update 해줘야 한다.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis', loc='right') # left, center, right 가능
plt.ylabel('Y-Axis', loc='top')   # botton, center, top 가능
plt.show()
```
