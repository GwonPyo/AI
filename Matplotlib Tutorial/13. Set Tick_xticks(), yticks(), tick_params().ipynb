<h3> <strong> 1. 기본 사용 </strong> </h3>

**xticks(), yticks() 함수**는 각각 x, y축에 눈금을 표시한다. <br>
xticks(), yticks() 함수에 파이썬 리스트 또는 numpy array를 입력하면 눈금과 숫자 레이블이 표시된다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo-')
plt.plot(x, x**2, 'r^-', linewidth=3)
plt.plot(x, x**3, 'gs-', markersize=9)
plt.xticks([0, 0.9, 1, 1.1, 2])
plt.yticks(np.arange(1, 6))

plt.show()
```

<h3> <strong> 2. 눈금 레이블 지정 </strong> </h3>

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo-')
plt.plot(x, x**2, 'r^-', linewidth=3)
plt.plot(x, x**3, 'gs-', markersize=9)
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.show()
```

<h3> <strong> 3. 눈금 스타일 설정 </strong> </h3>

* **tick_params()**: 눈금 스타일을 다양하게 설정할 수 있다.
* axis: 설정이 적용될 축을 지정한다. 'x', 'y', 'both' 중 선택할 수 있다.
* direction: 'in', 'out', 'inout' 중 선택할 수 있다.
* length: 눈금의 길이를 지정한다.
* pad: 눈금과 레이블과의 거리를 지정한다.
* labelsize: 레이블의 크기를 지정한다.
* labelcolor: 레이블의 색상을 지정한다.
* top/bottom/left/right: True로 지정하면 눈금이 표시될 위치를 선택할 수 있다.
* width: 눈금의 너비를 지정한다.
* color: 눈금의 색상을 지정한다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo-')
plt.plot(x, x**2, 'r^-', linewidth=3)
plt.plot(x, x**3, 'gs-', markersize=9)
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.tick_params(axis='x', direction='in', length=3, pad=6, labelsize=10, labelcolor='green', top=True)
plt.tick_params(axis='y', direction='inout', length=10, pad=15, labelsize=10, width=2, color='r')

plt.show()
```
