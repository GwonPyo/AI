<h3> <strong> 1. 기본 그래프 그리기 </strong> </h3>

**pyplot.plot()** 함수에 하나의 숫자 리스트를 입력하면 아래와 같은 그래프가 그려진다. <br>
해당 리스트의 값들을 y값으로 인식하고, x값은 [0, 1, 2, 3]으로 자동 생성된다. <br>
**show()** 함수는 그래프를 화면에 그려준다.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.show()
```

아래와 같이 2개의 인자를 입력받으면 첫 번째 인자는 x, 두 번째 인자는 y로 인식한다.

```python
import matplotlib.pyplot as plt

plt.plot([4, 3, 2, 1], [1, 2, 3, 4])
plt.show()
```

<h3> <strong> 2. 스타일 지정 </strong> </h3>

x, y 값 인자에 대해 선의 색상과 형태를 지정할 수 있다. 이러한 인자를 **Format String**이라고 한다. <br>
또한 axis를 이용해 축의 범위를 지정할 수 있다. ([xmin, xmax, ymin, ymax])

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ob')
plt.axis([0, 6, 0, 20])
plt.show()
```

<h3> <strong> 3. 여러 개의 그래프 </strong> </h3>

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 5, 0.2)

plt.plot(x, x, 'r--', x, x**2, 'bs', x, x**3, 'g^')
plt.show()
```
