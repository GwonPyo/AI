<h3> <strong> 1. 기본 사용 </strong> </h3>

np.arange()를 이용해 **좁은 간격으로 여러 정의역(x값)을 만들고 식에 대입**하면 곡선 그래프를 그릴 수 있다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'r-', x, x**2, 'bo:', x, x**3, 'g-.')
plt.show()
```

스타일을 지정하려면 선형과 동일하게 color, marker, linestyle로 지정할 수 있다.
