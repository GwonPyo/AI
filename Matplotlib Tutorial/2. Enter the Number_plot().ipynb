<h3> <strong> 1. 기본 사용 </strong> </h3>

**plot() 함수**에 하나의 리스트를 입력하면 y값으로 인식하고 그래프를 그린다.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4]) # plt.plot(np.array([1, 2, 3, 4])) 처럼 파이썬 튜플이나 Numpy array 형태로도 입력할 수 있다.
plt.show()             # x값은 기본으로 [0, 1, 2, 3]이 된다.
```

<h3> <strong> 2. x, y값 입력 </strong> </h3>

2개의 리스트를 입력하면 첫 번째 인자는 x, 두 번째 인자는 y로 인식한다.

```python
import matplotlib.pyplot as plt

plt.plot([4, 3, 2, 1], [1, 2, 3, 4])
plt.show()
```

<h3> <strong> 3. 레이블이 있는 데이터 사용 </strong> </h3>

dictionary type과 같이 레이블이 있는 데이터도 그래프로 나타낼 수 있다. <br>
아래처럼 레이블(key 값)을 입력해주고, data 파라미터에 딕셔너리를 지정해주면 된다.

```python
import matplotlib.pyplot as plt

data_dict = {'x': [1, 2, 3, 4, 5], 'y': [1, 10, 6, 5, 8]}

plt.plot('x', 'y', data=data_dict)
plt.show()
```
