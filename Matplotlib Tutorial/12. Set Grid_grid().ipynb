<h3> <strong> 1. 기본 사용 </strong> </h3>

**plt.grid(True)**로 설정하면, 그래프의 x, y축에 대해 그리드가 표시된다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo-')
plt.plot(x, x**2, 'r^-', linewidth=3)
plt.plot(x, x**3, 'gs-', markersize=9)
plt.grid(True)

plt.show()
```

<h3> <strong> 2. 축 지정 </strong> </h3>

**axis=y**로 설정하면 가로 방향의 그리드만 표시된다. <br>
'both', 'x', 'y' 중 하나로 지정하면 되고 default 값은 'both'다.


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo-')
plt.plot(x, x**2, 'r^-', linewidth=3)
plt.plot(x, x**3, 'gs-', markersize=9)
plt.grid(True, axis='y')

plt.show()
```

<h3> <strong> 3. 스타일 설정 </strong> </h3>

**color, alpha, linestyle**을 사용해 선의 스타일을 지정할 수 있다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo-')
plt.plot(x, x**2, 'r^-', linewidth=3)
plt.plot(x, x**3, 'gs-', markersize=9)
plt.grid(True, axis='y', linestyle='dashed', color='blue', alpha=0.5)

plt.show()
```
