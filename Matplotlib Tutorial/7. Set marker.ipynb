<h3> <strong> 1. 기본 사용 </strong> </h3>

plot() 함수에 'bo'를 입력해주면 파란색 원형 마커로 그래프가 표시된다.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1,4, 9, 16], 'bo')
plt.show()
```

<h3> <strong> 2. 선/마커 동시에 나타내기 </strong> </h3>

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'bo-')
plt.plot([1, 2, 3, 4], [1, 6, 15, 8], 'ro--')
plt.show()
```

<h3> <strong> 3. marker 파라미터 사용 </strong> </h3>

plot() 함수의 **marker 파라미터**를 사용하면 다양한 마커 형태를 지정할 수 있다.

```python
import matplotlib.pyplot as plt

plt.plot([4, 5, 6], marker="H")
plt.plot([3, 4, 5], marker="D")
plt.plot([2, 3, 4], marker="x")
plt.plot([1, 2, 3], marker='v')
plt.plot([0, 1, 2], marker='$O$')
plt.show()
```
