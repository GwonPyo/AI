<h3> <strong> 1. 기본 사용 </strong> </h3>

가장 기본적으로 plot()의 포맷 문자열을 사용해 선의 색상을 지정할 수 있다.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2.0, 3.0, 5.0, 10.0], 'r')
plt.plot([1, 2, 3, 4], [2.0, 2.8, 4.3, 6.5], 'g')
plt.plot([1, 2, 3, 4], [2.0, 2.5, 3.3, 4.5], 'b')
plt.show()
```

<h3> <strong> 2. color 키워드 인자 사용 </strong> </h3>

**color 파라미터**를 사용해 더 다양한 색상 이름을 지정할 수 있다.

* marker: 마커 스타일 / linestyle: 선 스타일 / color: 색상 지정

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2.0, 3.0, 5.0, 10.0], color='limegreen')
plt.plot([1, 2, 3, 4], [2.0, 2.8, 4.3, 6.5], color='violet')
plt.plot([1, 2, 3, 4], [2.0, 2.5, 3.3, 4.5], color='dodgerblue')
plt.show()
```

<h3> <strong> 3. Hex code 사용 </strong> </h3>

**16진수 코드(Hex code)**로 다양한 색상을 지정할 수 있다.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], color='#e35f62', marker='8', linestyle=':')
plt.show()
```
