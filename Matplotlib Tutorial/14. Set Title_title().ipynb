<h3> <strong> 1. 기본 사용 </strong> </h3>

**title() 함수**를 사용하면 그래프의 타이틀을 지정할 수 있다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo-')
plt.plot(x, x**2, 'r^-', linewidth=3)
plt.plot(x, x**3, 'gs-', markersize=9)

plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=1)
plt.title('Graph Title')

plt.show()
```

<h3> <strong> 2. 위치와 오프셋 지정 </strong> </h3>

**loc 파라미터**는 'left', 'right', 'center' 중 선택하여 title의 위치를 지정할 수 있다. <br>
**pad 파라미터**는 타이틀과 그래프의 간격을 설정할 수 있다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo-')
plt.plot(x, x**2, 'r^-', linewidth=3)
plt.plot(x, x**3, 'gs-', markersize=9)

plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=1)
plt.title('Graph Title', loc='right', pad=20)

plt.show()
```

<h3> <strong> 3. 폰트 지정 </strong> </h3>

**fonctdict 파리미터**에 딕셔너리 형태로 폰트 스타일을 설정할 수 있다. <br>
'fontsize'는 포인트 단위의 숫자를 입력하거나 'smaller', 'x-large' 등 상대적인 설정이 가능하다. <br>
'fontweight'에는 'normal', 'bold', 'heavy', 'light', 'ultrabold', 'ultralight'와 같이 설정할 수 있다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo-')
plt.plot(x, x**2, 'r^-', linewidth=3)
plt.plot(x, x**3, 'gs-', markersize=9)

plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=1)

title_font = {
    'fontsize': 16,
    'fontweight': 'bold'
}

plt.title('Graph Title', loc='right', pad=20, fontdict=title_font)

plt.show()
```

<h3> <strong> 4. 타이틀 얻기 </strong> </h3>

plt.title() 함수는 타이틀을 나타내는 **Matplotlib text 객체를 반환**한다. <br>
**get_position()**과 **get_text()** 메서드를 사용해 텍스트 위치와 문자열을 얻을 수 있다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo-')
plt.plot(x, x**2, 'r^-', linewidth=3)
plt.plot(x, x**3, 'gs-', markersize=9)

plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=1)
title_right = plt.title('Graph Title', loc='right', pad=20)

title_font = {
    'fontsize': 16,
    'fontweight': 'bold'
}

title_left = plt.title('Graph Title', loc='left', pad=20, fontdict=title_font)

print('title_right')
print(title_right.get_position())
print(title_right.get_text())
print('\n')

print('title_left')
print(title_left.get_position())
print(title_left.get_text())
print('\n')

plt.show()
```
