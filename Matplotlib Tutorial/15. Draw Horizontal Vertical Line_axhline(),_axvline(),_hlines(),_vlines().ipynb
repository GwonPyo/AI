<h3> <strong> 1. 수평선 그리기 - axhline(), hlines() </strong> </h3>

**axhline() 함수**의 첫 번째 인자는 y값으로 수평선의 위치가 된다. <br>
두, 세번째 인자는 xmin, xmax 값으로 0에서 1사이의 값을 입력한다. <br>
0은 왼쪽 끝, 1은 오른쪽 끝을 의미한다.

**hlines() 함수**는 y, xmin, xmax 값을 순서대로 입력하면 점 (xmin, y)에서 (xmax, y)를 따라 수평선을 표시한다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axhline(4.0, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=2)
plt.hlines(-0.62, 1.0, 2.5, color='gray', linestyle='solid', linewidth=3)

plt.show()
```

<h3> <strong> 2. 수직선 그리기 - axvline(), vlines() </strong> </h3>

**axvline() 함수**는 axhline()과 비슷하게 동작하고, **vlines() 함수**는 hlines() 함수와 비슷하게 동작한다. <br>
단지 x, ymin, ymax를 입력하는 차이밖에 없다.

```python
import matplotlib.pyplot as plt

x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axvline(1.0, 0.2, 0.8, color='lightgray', linestyle='--', linewidth=2)
plt.vlines(1.8, -3.0, 2.0, color='gray', linestyle='solid', linewidth=3)

plt.show()
```
