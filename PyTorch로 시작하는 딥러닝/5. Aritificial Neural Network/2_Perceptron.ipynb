## <strong> 1. 퍼셉트론 </strong>

**퍼셉트론(Perceptron)**은 초기 형태의 인공 신경망으로 다수의 입력으로부터 하나의 결과를 내보내는 알고리즘이다. <br>
실제 뇌를 구성하는 신경 세포 뉴런의 동작과 유사하다.

아래는 다수의 입력을 받는 퍼셉트론이다.

image.png

$x$는 입력값, $W$는 가중치, $y$는 출력값을 의미한다. <br>
이때 **가중치의 값이 클수록 입력 값이 중요하다는 것을 의미한다.**

이후 출력값이 **임계치(threshold)**를 넘으면 종착지에 있는 인공 뉴런은 출력 신호로 1을 출력하고 그렇지 않으면 0을 출력한다. <br>
이러한 함수를 **계단 함수(step function)**이라고 부른다.

```python
# step function
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-4, 4, 0.01)
y = (x > 0).astype('int64')

ax = plt.subplot(1, 1, 1)
ax.plot(x, y)
plt.show()
```

이때 임계치값을 수식으로 표현할 때는 $\theta$를 사용하며 계단 함수의 수식은 다음과 같다

$if \, \sum_{i}^{n}W_ix_i \geq \theta → y=1 $

$if \, \sum_{i}^{n}W_ix_i < \theta → y=0 $

이때 위 식에서 임계치를 좌변으로 넘기고 편향(bias)로 표현할 수 있다. <br>
수식으로는 아래와 같이 표현된다.

$if \, \sum_{i}^{n}W_ix_i+b \geq 0 → y=1 $

$if \, \sum_{i}^{n}W_ix_i+b < 0 → y=0 $


## <strong> 2. 단층 퍼셉트론 </strong>

위에서 배운 퍼셉트론을 **단층 퍼셉트론(Single-Layer Perceptron)**이라고 한다. <br>
단층 퍼셉트론은 값을 보내는 단계와 값을 출력하는 두 가지 단계로 이루어진다. <br>
이때 각 단계를 **층(layer)**라 부른다.

image.png

<h3> <strong> 2.1 AND, NAND, OR 게이트 </strong> </h3>

단층 퍼셉트론으로 AND, NAND, OR 게이트를 구현해보자.

```python
def AND_gate(x1, x2):
  w1 = 0.5
  w2 = 0.5
  b = -0.7
  result = w1*x1 + w2*x2 + b
  if result <= 0: return 0
  else: return 1

print(AND_gate(0, 0), AND_gate(1, 0), AND_gate(0, 1), AND_gate(1, 1))
```

```python
def NAND_gate(x1, x2):
  w1 = -0.5
  w2 = -0.5
  b = 0.7
  result = w1*x1 + w2*x2 + b
  if result <= 0: return 0
  else: return 1

print(NAND_gate(0, 0), NAND_gate(1, 0), NAND_gate(0, 1), NAND_gate(1, 1))
```

```python
def OR_gate(x1, x2):
    w1=0.6
    w2=0.6
    b=-0.5
    result = w1*x1 + w2*x2 + b
    if result <= 0: return 0
    else: return 1

print(OR_gate(0, 0), OR_gate(0, 1), OR_gate(1, 0), OR_gate(1, 1))
```

<h3> <strong> 2.2 XOR 게이트 </strong> </h3>

위처럼 단층 퍼셉트론으로 AND, NAND, OR 게이트를 구현할 수 있다. <br>
하지만 XOR 게이트는 구현할 수 없다. <br>
이는 **단층 퍼셉트론은 직선 하나로 두 영역을 나눌 수 있는 문제에 대해서만 구현이 가능하기 때문이다.**

OR, NAND 게이트는 아래와 같이 하나의 직선으로 두 영역을 나눌 수 있다.

하지만 XOR 게이트는 아래와 같이 하나의 직선으로 영역을 구분할 수 없다.

하지만 직선이 아닌 **곡선, 비선형 영역**으로 분리하면 구현이 가능하다.

## <strong> 3. 다층 퍼셉트론 </strong>

XOR 게이트는 기존의 AND, NAND, OR 게이트를 조합해 구현할 수 있다. <br>
즉, 여러 개의 층을 쌓으면 구현할 수 있다. <br>
단층 퍼셉트론은 입력층과 출력층만 존재했지만 다층 퍼셉트론은 사이에 **은닉층(hidden layer)**가 존재한다. <br>
